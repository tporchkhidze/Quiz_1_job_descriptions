```python
import pandas as pd
df = pd.read_csv('job_descriptions.csv')
df[["Experience", "Job Title", "skills", "Salary Range"]].head(20)
```

ინდექსირება "Job Id"-ის მიმართ

```python
df = pd.read_csv("job_descriptions.csv", index_col="Job Id")
df.head(20)
```

საქართველოს მონაცემების მიხედვით, სადაც Company Size მეტია 50000-ზე

```python
country = ["Georgia"]
filtered_company_size = df['Company Size']>50000
filtered_country = df['Country'].isin(country)
df.loc[filtered_company_size & filtered_country, ["Company Size",'Country',  'Role',"Salary Range"]].head(20)
```

სორტირება

```python
df.sort_values(by=['Job Posting Date', 'Job Title'], ascending=[True, False], inplace=True)
df[['Job Posting Date', 'Job Title']].head(20)
```

სტატისტიკური ფუნქციები Company Size-ის მნიშვნელობებზე

```python
# Company Size.min
df["Company Size"].min()
```

```python
# Company Size.max
df["Company Size"].max()
```

```python
# Company Size.mean
df["Company Size"].mean()
```

```python
# Company Size.median
df["Company Size"].median()
```

```python
# Company Size.std
df["Company Size"].std()
```

სვეტოვანი დიაგრამა

```python
import matplotlib.pyplot as plt

x = df["Country"].head(10)
y = df["Salary Range"].head(10)

fig, ax = plt.subplots()
ax.barh(x, y, color = "red")

plt.xticks(rotation='vertical')

ax.set_xlabel('Salary Range')
ax.set_ylabel('Countries')

ax.set_title('Top 10 Countries by Salary Range')

plt.show()
```

წრიული დიაგრამა


```python
sizes = df["Company Size"].head(10)
labels = df["Company"].head(10)

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%.0f%%')

ax.set_title('Company Sizes for Top 10 Companies')

plt.show()
```
