ინდექსირება "ResponseId"-ის მიმართ

```python
df = pd.read_csv("survey_results_public.csv", index_col="ResponseId")
df.head(20)
```

საქართველოს მონაცემების მიხედვით, სადაც CompTotal მეტია 50000-ზე

```python
country = ["Georgia"]
filtered_company_size = df['CompTotal']>50000
filtered_country = df['Country'].isin(country)
df.loc[filtered_company_size & filtered_country, ["Country", "CompTotal", "Employment", "YearsCode", "EdLevel"]].head(20)
```

სორტირება

```python
df.sort_values(by=['OrgSize', 'Age'], ascending=[True, False], inplace=True)
df[['OrgSize', 'Age']].head(50)
```

სვეტოვანი დიაგრამა

```python
import matplotlib.pyplot as plt

x = df["DevType"].head(10)
y = df["YearsCode"].head(10)

fig, ax = plt.subplots()
ax.barh(x, y, color = "red")

plt.xticks(rotation='vertical')

ax.set_xlabel('DevType')
ax.set_ylabel('YearsCode')

ax.set_title('Top 10 Developer Types and Years of Coding Experience')

plt.show()
```

წრიული დიაგრამა


```python
sizes = df["YearsCode"].head(10)
labels = df["Country"].head(10)

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%.0f%%')

ax.set_title('Distribution of Coding Experience by Country')

plt.show()
```
